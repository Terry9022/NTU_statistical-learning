```python
#根據上面手寫題 定義 Gradient 和 Hessian 以及 Error function
def Gradient(w, big_lambda, x, y, t):
    return big_lambda.dot(w) + x.T.dot(y-t)
def Hessian(w, big_lambda, x, y, t):
    n = len(y)
    R = np.zeros((n,n))
    for i in range(n):
        R[i][i] = y[i][0] * (1 - y[i][0])
    return big_lambda + x.T.dot(R).dot(x)
def Error(w, y, t):
    loss = 0 
    for i in range(len(y)):
        loss += -( t[i][0] * np.log(y[i][0]) + (1-t[i][0]) * np.log( 1 - y[i][0]))
    return loss
```

```python
#定義 增加常數項函數 以及 sigmoid函數
def add_intercept(X):
    intercept = np.ones((X.shape[0], 1))
    return np.concatenate((X,intercept), axis=1)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
#mylogistic_l2 class
class mylogistic_l2:

    def __init__(self, reg_vec, max_iter, tol, add_intercept):
        self.reg_vec = reg_vec
        self.feature_dim = len(reg_vec)
        self.big_lambda = np.diag(self.reg_vec)     # big_lambda is The big lambda matrix   
        self.max_iter = max_iter                    # max_iter : The maximum number of iterations
        self.tol = tol                              # tolerance
        self.add_intercept = add_intercept          # whether add intercept
        
    def fit(self, x_train, y_train):
        self.num = x_train.shape[0]
        self.x = x_train                                 # shape = (num, feature_dim)
        self.t = y_train.reshape((1,len(y_train))).T     # shape = (num, 1)
        self.b = np.average(self.reg_vec)
        
        if self.add_intercept:
            self.x=add_intercept(X=self.x)              # shape = (num, feature_dim + 1)
            
        self.w = np.linalg.inv((self.x.T).dot(self.x) + (self.b)*np.identity(self.feature_dim)) \
        .dot(self.x.T).dot(self.t)                       # shape = (deature_dim, 1)
        
        self.y = np.zeros((self.num,1))                  # shape = (num, 1)
        for i in range(self.num):
            self.y[i][0] = sigmoid(self.w.T.dot(self.x[i]))
        
        iter_time = 0                                   # count number of iterations
        
        while (iter_time <= self.max_iter):
            cur_loss = Error(self.w, self.y, self.t)
            print("Iteration #:  {:>7,.0f}.  Cost: {:>+7.4f}.".format(iter_time,cur_loss))  # print current cost
            grad = Gradient(self.w, self.big_lambda, self.x, self.y, self.t)   # Gradient
            hess = Hessian(self.w, self.big_lambda, self.x, self.y, self.t)    # Hessian
            new_w = self.w - np.linalg.inv(hess).dot(grad)    # 牛頓法公式 更新新的weight
            
            for i in range(self.num):
                self.y[i][0] = sigmoid(new_w.T.dot(self.x[i]))  # update y
            new_loss = Error(new_w, self.y, self.t)
           
            self.w = new_w    # update w
            
        
            if np.abs(new_loss - cur_loss) < self.tol:       # 當誤差達到我們要求時 便停止 不再更新 而且紀錄 weight
                print()
                print("cost function tolerance reached")      # print final cost
                print("Iteration #:  {:>7,.0f}. Final Cost: {:>+7.4f}.".format(iter_time,cur_loss))
                break
            
            iter_time += 1
    
    

    def predict_prob(X,theta,fitIntercept=True):
        if fitIntercept:
            X = add_intercept(X=X)
        
        return sigmoid(np.dot(X, theta))   #將結果丟入sigmoid函數 吐出機率值

    def predict(self, X):
        predictedProb = mylogistic_l2.predict_prob(X, self.w, self.add_intercept)
        predictedClasses = [0 if x <= 0.5 else 1 for x in predictedProb]
        return np.array(predictedClasses)   # 如果機率超過0.5 則預測結果為1 回傳預測值 0或是1
```

```python
#Computer binary classification performance metrics
def performanceEval(predictions, y_test):   
        #Initialize
        TP, TN, FP, FN, P, N = 0, 0, 0, 0, 0, 0        
        for idx, test_sample in enumerate(y_test):
            
            if predictions[idx] == 1 and test_sample == 1:
                TP += 1       
                P += 1
            elif predictions[idx] == 0 and test_sample == 0:                
                TN += 1
                N += 1
            elif predictions[idx] == 0 and test_sample == 1:
                FN += 1
                P += 1
            elif predictions[idx] == 1 and test_sample == 0:
                FP += 1
                N += 1
            
        accuracy = (TP + TN) / (P + N)            # definition of accuracy    
        return accuracy
```

```python
# Split the training data into subtraining (90%) and tuning (10%) to search for the best hyperparameters
x_train_train=x_train[:-3016]
x_train_tune=x_train[-3016:]
```

```python
y_train_train=y_train[:-3016]
y_train_tune=y_train[-3016:]
```

```python
#10 grids in [0.01, 100]
grids=[0.01,0.025,0.05,0.1,0.5,1,5,10,50,100]
```

```python
# Conduct grid search with the constraint that  𝑎1=𝑎2 . Record the best value  𝑎1∗  and  𝑎2∗
accuracy1=[]
for i in grids:
    a1a2=[i]*x_train_train.shape[1]
    vec = np.r_[a1a2,0]
    logic = mylogistic_l2(vec,1000,1e-5, True)
    logic.fit(x_train_train,y_train_train)
    pred=logic.predict(x_train_tune)
    accuracy1.append(performanceEval(pred,y_train_tune))
```

```python
# Plot accuracy and lambda
line1, = plt.plot(grids, accuracy1, "blue", label= "Case1:")
plt.legend()
plt.xlabel("Lambda")  
plt.ylabel("Accuracy")  
plt.show() 
```

可以看到最好的 Lambda 為5

```python
accuracy1
```

```python
#Fix  𝑎1=5 , and search  𝑎2  for the best value, call the result the new  𝑎2∗ .
accuracy2=[]
for i in grids:
    a1=[5]*6
    a2=[i]*(x_train_train.shape[1]-6)
    vec = np.r_[a1,a2,0]
    logic = mylogistic_l2(vec,1000,1e-5, True)
    logic.fit(x_train_train,y_train_train)
    pred=logic.predict(x_train_tune)
    accuracy2.append(performanceEval(pred,y_train_tune))
```

```python
# Plot accuracy and lambda
line1, = plt.plot(grids, accuracy2, "blue", label= "Case2:")
plt.legend()
plt.xlabel("Lambda")  
plt.ylabel("Accuracy")  
plt.show() 
```
